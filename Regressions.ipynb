# Linear Regression
Load the `housing` dataset. Split it into training and test sets. Train and test a linear regression model using scikit-learn. Check the documentation to identify the most important hyperparameters, attributes, and methods of the model. Use them in practice.

## Importing Modules

```python
import pandas as pd
import sklearn.model_selection
import sklearn.linear_model
import sklearn.metrics
import sklearn.preprocessing
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the Dataset

```python
df = pd.read_csv('housing.csv')
df = df.set_index("id")

df
# df.plot(kind='scatter', x='bedrooms', y='price')
# plt.show()
```

## Spliting Dataset  train and test datasets

```python
x = df.drop(['price'], axis=1)
y = df['price']


x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y)
# print('df: ', df.shape)
print('x_train: ', x_train)
# print('x_test: ', x_train.shape)
# print('y_train: ', y_train.shape)
# print('y_test: ', y_test.shape)
```

## Training a Model

```python
model = sklearn.linear_model.LinearRegression()
model.fit(x_train, y_train)
```

## Testind the trained Model

```python
y_predicted = model.predict(x_test)
mse = sklearn.metrics.mean_squared_error(y_test, y_predicted)
mse
```

## Polynomial Regression

```python
poly = sklearn.preprocessing.PolynomialFeatures(degree=2)
poly.fit(x_train)

x_train_new = poly.transform(x_train)
# x_train_new.shape

# Training another linear regression model
model = sklearn.linear_model.LinearRegression()
model.fit(x_train_new, y_train)

x_test_new = poly.transform(x_test)
y_predicted = model.predict(x_test_new)
mse = sklearn.metrics.mean_squared_error(y_test, y_predicted)
mse
```
